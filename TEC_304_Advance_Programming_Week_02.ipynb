Programming paradigms are different styles of writing programs.


* **Procedural programming**: code is written as step-by-step procedures or functions.
* **Object-oriented programming (OOP)**: code is organized around objects that combine data and behavior.
* **Functional programming**: focuses on pure functions, immutability, and transforming data.
* **Logic programming**: you describe rules and facts, and the system figures out answers.
* **Event-driven programming**: program flow depends on events like clicks, input, or messages.
* **Declarative programming**: you describe **what** you want, not **how** to do it.
* **Linear programming**: a mathematical optimization method, not a usual coding style; it finds the best solution under linear constraints.



````python
# Small examples of programming paradigms

# 1. Procedural
def add(a, b):
    return a + b

print("Procedural:", add(2, 3))


# 2. Object-Oriented Programming (OOP)
class Dog:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f"{self.name} says woof"

d = Dog("Max")
print("OOP:", d.speak())

# 3. Functional
nums = [1, 2, 3, 4]
squares = list(map(lambda x: x * x, nums))
print("Functional:", squares)

# 4. Declarative style
evens = [x for x in nums if x % 2 == 0]
print("Declarative:", evens)

# 5. Event-driven
def on_click():
    print("Button clicked")

on_click()   # simulating an event


# 6. Logic programming idea (Python imitation, real logic programming is often Prolog)
facts = {"human": ["Ali"], "mortal": []}
if "Ali" in facts["human"]:
    facts["mortal"].append("Ali")
print("Logic-style:", "Ali is mortal" if "Ali" in facts["mortal"] else "Unknown")


# 7. Linear programming idea
# Maximize profit subject to constraints.
# Usually solved with libraries like scipy or pulp.
print("Linear Programming: optimize an objective like max profit under constraints")
````

````python
count = 0   # global variable

def add_one(x):
    global count
    count += 1          # side effect: changes external state
    print("Called")     # side effect: output to screen
    return x + 1        # main return value

result = add_one(5)

print(result)   # 6
print(count)    # 1


def add(a, b):
    return a + b

result = add(3, 4)
print(result)   # 7

````

Pure function: just returns a result

\
Non-pure function: changes something outside the functio

````python
# Pure function
def multiply_by_2(x):
    return x * 2

print(multiply_by_2(5))   # 10


# Non-pure function
numbers = []

def save_number(x):
    numbers.append(x)     # changes external list
    return x * 2

print(save_number(5))     # 10
print(numbers)            # [5]
````

Functional Programming

````python
# Calculator using built-in pure functions

a = 10
b = 5

print(abs(-7))         # absolute value -> 7
print(pow(a, b))       # 10^5 -> 100000
print(max(a, b))       # larger value -> 10
print(min(a, b))       # smaller value -> 5
print(round(3.14159, 2))  # rounded -> 3.14
print(sum([a, b, 2]))  # total -> 17
````

These are common **functional programming characteristics**:

**1. More declarative instead of imperative**
You describe **what** result you want, not every step of **how** to do it.

* **Imperative:** “make an empty list, loop, add items”
* **Declarative:** “give me the squares”

```python
nums = [1, 2, 3, 4]

# Declarative style
squares = list(map(lambda x: x * x, nums))
print(squares)   # [1, 4, 9, 16]
```

---

**2. Functions as first-class objects and higher-order functions**
Functions can be:

* stored in variables
* passed as arguments
* returned from other functions

A **higher-order function** is a function that takes another function as input or returns one.

```python
def double(x):
    return x * 2

def apply_function(f, value):
    return f(value)

print(apply_function(double, 5))   # 10
```

Here, `double` is passed like a normal value.

---

**3. Limited or no side effects**
A function should mainly return a value, without changing global variables, printing, or modifying outside data.

```python
def add(a, b):
    return a + b

print(add(2, 3))   # 5
```

This is clean because it only computes and returns a result.

---

**4. Immutability of objects**
Data is not changed after creation. Instead of modifying it, you create new data.

```python
numbers = (1, 2, 3)      # tuple is immutable
new_numbers = numbers + (4,)

print(numbers)      # (1, 2, 3)
print(new_numbers)  # (1, 2, 3, 4)
```

The original object stays unchanged.

---

**Simple summary**

* **Declarative** = say what you want
* **First-class functions** = treat functions like values
* **No side effects** = avoid changing outside state
* **Immutability** = create new data instead of changing old data

Lambda fuction

````python
# 1. No argument
greet = lambda: "Hello"
print(greet())          # Hello

# 2. One argument
square = lambda x: x * x
print(square(4))        # 16

# 3. Two arguments
add = lambda a, b: a + b
print(add(3, 5))        # 8

# 4. Three arguments
total = lambda a, b, c: a + b + c
print(total(1, 2, 3))   # 6

# 5. With condition
bigger = lambda a, b: a if a > b else b
print(bigger(10, 7))    # 10

# 6. Used inside another function
nums = [1, 2, 3, 4]
squares = list(map(lambda x: x * x, nums))
print(squares)          # [1, 4, 9, 16]

is_even = lambda x: x % 2 == 0

print(is_even(4))   # True


is_even2 = lambda a: "EVEN" if a % 2== 0 else "ODD"
print(is_even2(11 ))    # 10
````

MAP, reduce, and filter

````python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

# map: change every value
squares = list(map(lambda x: x * x, numbers))
print(squares)   # [1, 4, 9, 16, 25]

# filter: keep only matching values
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)     # [2, 4]

# reduce: combine all values into one value
total = reduce(lambda a, b: a + b, numbers)
print(total)     # 15
````
