| Family         | Meaning                                |                     Default? | Address format / example     | Purpose                                                         | Common use case                                                      |
| -------------- | -------------------------------------- | ---------------------------: | ---------------------------- | --------------------------------------------------------------- | -------------------------------------------------------------------- |
| `AF_INET`      | IPv4 Internet addresses                | Yes, most common in teaching | `("127.0.0.1", 5000)`        | Network communication using IPv4                                | Client-server apps, web apps, chat apps, file transfer               |
| `AF_INET6`     | IPv6 Internet addresses                |                           No | `("::1", 5000)`              | Network communication using IPv6                                | Modern networks, future-proof systems, large-scale internet services |
| `AF_UNIX`      | Unix local socket                      |                           No | `"/tmp/mysocket"`            | Communication between processes on the same machine             | Local IPC on Linux/macOS, fast communication between services        |
| `AF_CAN`       | Controller Area Network                |                           No | CAN interface and identifier | Communication in CAN bus systems                                | Cars, industrial controllers, embedded systems                       |
| `AF_RDS`       | Reliable Datagram Sockets              |                           No | Specialized addressing       | Reliable datagram-based communication                           | Specialized cluster/network systems                                  |
| `AF_PACKET`    | Direct packet-level interface on Linux |                           No | Network interface level      | Access raw packets at link layer                                | Packet sniffing, network analysis, security tools                    |
| `AF_BLUETOOTH` | Bluetooth communication                |                           No | Bluetooth device addresses   | Wireless communication over Bluetooth                           | IoT devices, wireless peripherals                                    |
| `AF_NETLINK`   | Kernel-user communication on Linux     |                           No | Kernel interface             | Communication between user programs and Linux kernel subsystems | System/network management tools                                      |


| Socket type      | Meaning                     | Default in examples? | Usually maps to                        | Communication style                               | Purpose                                      | Common use case                                   |
| ---------------- | --------------------------- | -------------------: | -------------------------------------- | ------------------------------------------------- | -------------------------------------------- | ------------------------------------------------- |
| `SOCK_STREAM`    | Stream socket               |                  Yes | TCP                                    | Connection-oriented, reliable, ordered            | Continuous data stream between two endpoints | Web servers, chat systems, client-server apps     |
| `SOCK_DGRAM`     | Datagram socket             |                   No | UDP                                    | Connectionless, faster, not guaranteed            | Send separate packets/messages               | DNS, video streaming, online games                |
| `SOCK_RAW`       | Raw socket                  |                   No | Raw IP / packet access                 | Very low-level access                             | Manual packet handling                       | Security tools, packet sniffers, protocol testing |
| `SOCK_SEQPACKET` | Sequential packet socket    |                   No | Depends on family                      | Connection-oriented, preserves message boundaries | Reliable message-based communication         | Advanced IPC / special OS-level systems           |
| `SOCK_RDM`       | Reliably delivered messages |                   No | Rarely used                            | Message-based reliable delivery                   | Specialized reliable messaging               | Rare / advanced networking                        |
| `SOCK_NONBLOCK`  | Non-blocking flag           |                   No | Flag, not really a main type by itself | Operations do not wait                            | Faster async-style behavior                  | High-performance servers                          |
| `SOCK_CLOEXEC`   | Close-on-exec flag          |                   No | Flag                                   | Auto close on process exec                        | Resource safety                              | Secure system programming                         |


| Protocol value / constant | Meaning                               | Usually used with     | Purpose                         | Use case                             |
| ------------------------- | ------------------------------------- | --------------------- | ------------------------------- | ------------------------------------ |
| `0`                       | Default protocol chosen automatically | Any family/type combo | Simplest and most common choice | Almost all beginner programs         |
| `IPPROTO_TCP`             | TCP                                   | `SOCK_STREAM`         | Reliable transport              | Web, chat, client-server apps        |
| `IPPROTO_UDP`             | UDP                                   | `SOCK_DGRAM`          | Fast packet-based transport     | Streaming, games, DNS                |
| `IPPROTO_ICMP`            | ICMP                                  | Often raw sockets     | Control/error messages          | Ping tools, diagnostics              |
| `IPPROTO_RAW`             | Raw IP                                | `SOCK_RAW`            | Manual packet creation          | Security tools, protocol experiments |
| `IPPROTO_IPV6`            | IPv6 protocol layer                   | IPv6 sockets          | IPv6-specific handling          | Advanced IPv6 networking             |


| Method       | Purpose                                                            | How it works                                                                               | Typical use case                                                 | TCP / UDP  | Student-friendly explanation                                             |
| ------------ | ------------------------------------------------------------------ | ------------------------------------------------------------------------------------------ | ---------------------------------------------------------------- | ---------- | ------------------------------------------------------------------------ |
| `connect()`  | Establishes a connection to a remote socket                        | The client specifies the server address and port, and the OS performs the connection setup | Client connecting to a web server, chat server, or file server   | Mostly TCP | Used by the client to start communication with a server                  |
| `send()`     | Sends data through a connected socket                              | Sends some bytes to the other side, but it may not send all bytes in one call              | Sending a short message over an existing TCP connection          | Mostly TCP | Sends data, but you may need to call it again if everything was not sent |
| `sendall()`  | Sends all data through a connected socket                          | Keeps sending until all bytes are transmitted or an error occurs                           | Sending complete files, messages, or long data reliably over TCP | Mostly TCP | Safer than `send()` when you want the whole message to go                |
| `sendto()`   | Sends data to a specific address without first making a connection | The sender gives both the data and destination address in the same call                    | UDP messaging, DNS queries, simple broadcast-style communication | Mostly UDP | Used when you want to send a packet directly to a target address         |
| `recv()`     | Receives data from a connected socket                              | Reads up to a specified number of bytes from the socket buffer                             | Receiving replies or messages in TCP client-server communication | Mostly TCP | Used to read incoming data from a connected socket                       |
| `recvfrom()` | Receives data and also tells you where it came from                | Returns both the received data and the sender’s address                                    | UDP servers, datagram-based communication                        | Mostly UDP | Useful when messages may come from different senders                     |
